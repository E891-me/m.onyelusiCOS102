```python
import tkinter as tk
from tkinter import messagebox

class DeliveryCalculator:
    """Class to handle the logic of price calculation."""
    def calculate_price(self, location, weight):
        # Pricing logic based on the prompt requirements
        if location == "PAU":
            if weight >= 10:
                return 2000
            else:
                return 1500
        elif location == "Epe":
            if weight >= 10:
                return 5000
            else:
                return 4000
        else:
            return None

class DeliveryApp:
    """Class to handle the GUI and user interaction."""
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Pricing System")
        self.root.geometry("350x300")
        
        # Instance of the logic class
        self.calculator = DeliveryCalculator()

        # Location Selection
        tk.Label(root, text="Select Location:", font=("Arial", 10)).pack(pady=5)
        self.location_var = tk.StringVar(value="PAU")
        tk.Radiobutton(root, text="PAU", variable=self.location_var, value="PAU").pack()
        tk.Radiobutton(root, text="Epe", variable=self.location_var, value="Epe").pack()

        # Weight Input
        tk.Label(root, text="Package Weight (kg):", font=("Arial", 10)).pack(pady=10)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        # Calculate Button
        self.calc_button = tk.Button(root, text="Calculate Cost", command=self.get_cost, 
                                     bg="green", fg="white", font=("Arial", 10, "bold"))
        self.calc_button.pack(pady=20)

    def get_cost(self):
        try:
            # Get data from UI
            location = self.location_var.get()
            weight = float(self.weight_entry.get())
            
            # Use logic class to get price
            price = self.calculator.calculate_price(location, weight)
            
            if price is not None:
                messagebox.showinfo("Quote", f"Delivery to {location}\nWeight: {weight}kg\nTotal Cost: ₦{price}")
            else:
                messagebox.showerror("Error", "Invalid Location selected.")
                
        except ValueError:
            messagebox.showerror("Input Error", "Please enter a valid number for weight.")

# Main loop
if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryApp(root)
    root.mainloop()
```
