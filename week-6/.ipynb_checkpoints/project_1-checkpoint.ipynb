```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        # Data provided in the prompt
        self.employee_list = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

    def check_employee(self, name):
        """Checks if the name exists in the company records."""
        if name in self.employee_list:
            return True
        return False

    def take_attendance(self, name):
        """Simulates taking attendance."""
        return f"Attendance recorded for {name}."

    def assign_task(self):
        """Randomly selects a task using random.randint."""
        index = random.randint(0, len(self.tasks) - 1)
        return self.tasks[index]

    def refuse_access(self):
        """Returns a refusal message for unauthorized users."""
        return "Access Denied: You are not recognized as an employee."

class EmployeeApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Mrs. Jane's Logistics System")
        self.root.geometry("400x250")
        
        # Instance of the logic class
        self.logic = Employee()

        # UI Elements
        self.label = tk.Label(root, text="Enter Employee Name:", font=("Arial", 12))
        self.label.pack(pady=10)

        self.name_entry = tk.Entry(root, font=("Arial", 12), width=30)
        self.name_entry.pack(pady=5)

        self.login_button = tk.Button(root, text="Login / Check In", command=self.process_login, 
                                      bg="blue", fg="white", font=("Arial", 10, "bold"))
        self.login_button.pack(pady=20)

    def process_login(self):
        user_input = self.name_entry.get().strip()
        
        if self.logic.check_employee(user_input):
            # Sequence for successful login
            attendance_msg = self.logic.take_attendance(user_input)
            task = self.logic.assign_task()
            
            messagebox.showinfo("Success", f"{attendance_msg}\n\nYour task today: {task}")
        else:
            # Sequence for failed login
            refusal_msg = self.logic.refuse_access()
            messagebox.showerror("Unauthorized", refusal_msg)

# Main execution
if __name__ == "__main__":
    root = tk.Tk()
    app = EmployeeApp(root)
    root.mainloop()
```
