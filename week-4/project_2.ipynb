```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/DELL/OneDrive/Documents/m.onyelusiCOS102/titanic.csv")
df
```

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isna().sum()
```

```python
df
```

```python
plt.figure(figsize=(10, 8))
# Filter for numeric columns only
numeric_df = df.select_dtypes(include=['float64', 'int64']).drop(columns=['PassengerId'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
plt.close()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Survived', y='Age', data=df, palette='Set1', split=True)
plt.title('Age Distribution by Survival Status')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.savefig('age_violin_survival.png')
plt.close()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='FamilySize', y='Survived', data=df, marker='o', color='coral')
plt.title('Survival Rate vs Family Size')
plt.savefig('survival_by_family_size.png')
plt.close()
```
