```python
plt.hist(df['price'], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df['room_type'].value_counts().plot(kind='bar')
plt.title("Room Type Distribution")
plt.show()
```
