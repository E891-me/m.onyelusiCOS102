```python
# declare the variables and get input from user
print("SIMPLE INTEREST CALCULATOR")
p = float(input("Enter the amount: "))
r = float(input("Ente the interest rate: "))
t = float(input("Enter the amount of years: "))

# calculating for simple interest
a = p * (1 + (r / 100) * t)

#displaying the ouput of simple interest
print(f"The simple interest is: {a:.2f}")
```
