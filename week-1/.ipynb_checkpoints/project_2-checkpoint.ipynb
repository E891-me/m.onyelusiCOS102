```python
# declare the variables and get input from the user
p = float(input("Enter the principal amount: "))
r = float(input("Enter the interest rate: "))
n = float(input("Enter the number of times the interest is compounded per year: "))
t = float(input("Enter the amount of years: "))

# Calculating for the compound interest
a = p * (1 + r / n) ** (n * t)

#displaying the compound interest
print(f"The Compound Interest is {a:.2f}")
```
