```python
import tkinter as tk
from tkinter import messagebox

# --- OOP Logic ---

class Zenith:
    """Parent Class representing the Bank"""
    def __init__(self, employee_name):
        self.employee_name = employee_name

    def unique_services(self):
        return "Generic Zenith Unique Services"

    def mutual_services(self):
        return "Generic Zenith Mutual Services"

class Retail(Zenith):
    """Subclass for Retail Banking"""
    def unique_services(self):
        return ("• Retirement and education accounts\n"
                "• Loans and mortgages\n"
                "• Checking and saving")

    def mutual_services(self):
        return ("• Lines of credit\n"
                "• Investment management\n"
                "• Insurance")

class Global(Zenith):
    """Subclass for Global Banking (Polymorphism in action)"""
    def unique_services(self):
        return ("• Multi-currency management\n"
                "• Foreign currency accounts\n"
                "• Foreign currency credit cards\n"
                "• Transborder advisory services")

    def mutual_services(self):
        return "• Liquidity management"

class Commercial(Zenith):
    """Subclass for Commercial Banking"""
    def unique_services(self):
        return "• Advisory services"

    def mutual_services(self):
        return ("• Lines of credit\n"
                "• Investment management\n"
                "• Insurance")

# --- GUI Logic ---

class ZenithApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank Division Portal")
        self.root.geometry("400x450")

        # Labels and Inputs
        tk.Label(root, text="Employee Name:", font=("Arial", 10, "bold")).pack(pady=5)
        self.ent_name = tk.Entry(root)
        self.ent_name.pack(pady=5)

        tk.Label(root, text="Select Division:", font=("Arial", 10, "bold")).pack(pady=5)
        self.division_var = tk.StringVar(value="Retail")
        divisions = ["Retail", "Global", "Commercial"]
        self.drop_menu = tk.OptionMenu(root, self.division_var, *divisions)
        self.drop_menu.pack(pady=5)

        # Action Button
        self.btn_search = tk.Button(root, text="Show Services", command=self.display_info, bg="#E21E26", fg="white")
        self.btn_search.pack(pady=20)

        # Display Area
        self.lbl_result_title = tk.Label(root, text="", font=("Arial", 10, "italic"))
        self.lbl_result_title.pack()
        
        self.txt_display = tk.Text(root, height=10, width=40, state='disabled', bg="#f4f4f4")
        self.txt_display.pack(pady=10)

    def display_info(self):
        name = self.ent_name.get()
        division = self.division_var.get()

        if not name:
            messagebox.showwarning("Input Error", "Please enter an employee name.")
            return

        # Instantiate based on division (Polymorphism)
        if division == "Retail":
            emp_obj = Retail(name)
        elif division == "Global":
            emp_obj = Global(name)
        else:
            emp_obj = Commercial(name)

        # Update Display
        self.lbl_result_title.config(text=f"Employee: {name} | Division: {division}")
        
        self.txt_display.config(state='normal')
        self.txt_display.delete('1.0', tk.END)
        self.txt_display.insert(tk.END, f"--- Unique Services ---\n{emp_obj.unique_services()}\n\n")
        self.txt_display.insert(tk.END, f"--- Mutual Services ---\n{emp_obj.mutual_services()}")
        self.txt_display.config(state='disabled')

if __name__ == "__main__":
    root = tk.Tk()
    app = ZenithApp(root)
    root.mainloop()
```
